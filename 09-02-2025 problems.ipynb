```python
import pandas as pd
df = pd.DataFrame({
    'Product': ['Laptop', 'Tablet', 'Laptop', 'Smartphone'],
    'Price': [1200, 600, 1100, 900],
    'Stock': [30, 50, 20, 80]
})
df['Product'] = df['Product'].astype('category')
df['Price'] = pd.to_numeric(df['Price'], downcast='integer')
df['Stock'] = pd.to_numeric(df['Stock'], downcast='integer')
print(df.info())
```

problem 2

```python
import pandas as pd

index = pd.MultiIndex.from_tuples([
    ('Electronics', 'New York'),
    ('Electronics', 'San Francisco'),
    ('Furniture', 'New York'),
    ('Furniture', 'Los Angeles'),
    ('Clothing', 'New York'),
    ('Clothing', 'Chicago')
], names=['Category', 'City'])

data = {
    'Product': ['Laptop', 'Tablet', 'Sofa', 'Desk', 'T-shirt', 'Jeans'],
    'Price': [1200, 600, 300, 150, 20, 40],
    'Stock': [30, 50, 20, 15, 100, 80]
}

df = pd.DataFrame(data, index=index)

new_york_products = df.xs('New York', level='City')

print(new_york_products)
```

```python
import pandas as pd

df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Salary': [50000, 60000, 70000]})

result = pd.merge(df1, df2, on='ID', how='outer')

print(result)
```

problem 4

```python
import pandas as pd

df = pd.DataFrame({
    'Region': ['North', 'South', 'North', 'South'],
    'Product': ['A', 'B', 'A', 'B'],
    'Sales': [100, 200, 150, 250]
})

pivot_table = df.pivot_table(values='Sales', index='Region', columns='Product', aggfunc='sum')

print(pivot_table)
```

problem 5

```python
import pandas as pd

df = pd.DataFrame({
    'Region': ['East', 'East', 'West', 'West'],
    'Sales': [100, 150, 200, 250]
})

sales_range = df.groupby('Region')['Sales'].agg(lambda x: x.max() - x.min())

print(sales_range)
```
