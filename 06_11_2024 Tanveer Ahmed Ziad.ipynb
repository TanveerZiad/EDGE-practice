```python
stack = []

def push(element):
    stack.append(element)
    print(f"{element} pushed onto the stack.")

def pop():
    if len(stack) == 0:
        print("Error: Cannot pop from an empty stack.")
    else:
        element = stack.pop()
        print(f"{element} popped from the stack.")

def display():
    print("Current stack:", stack)

while True:
    print("1. Push an element onto the stack")
    print("2. Pop an element from the stack")
    print("3. Display the current stack")
    print("4. Exit")

    choice = input("Enter your choice (1/2/3/4): ")

    if choice == '1':
        element = input("Enter the element to push: ")
        push(element)
    elif choice == '2':
        pop()
    elif choice == '3':
        display()
    elif choice == '4':
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please try again.")
```

```python
import random
random_numbers = [random.randint(1, 100) for _ in range(10)]
print("List of random numbers: ", random_numbers)

max_value = max(random_numbers)
print("Maximum value: ", max_value)

min_value = min(random_numbers)
print("Minimum value: ", min_value)

average_value = sum(random_numbers) / len(random_numbers)
print("Average value: ", average_value)
```
