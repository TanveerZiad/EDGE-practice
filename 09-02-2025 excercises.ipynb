```python
import pandas as pd

df = pd.DataFrame({
    'Category': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Apple'],
    'Value': [10, 20, 30, 40, 50, 60]
})

print("Before optimization:")
print(df.dtypes)

df['Category'] = df['Category'].astype('category')

print("\nAfter optimization:")
print(df.dtypes)
```

```python
import pandas as pd

arrays = [
    ['USA', 'USA', 'Canada', 'Canada'],
    ['East', 'West', 'East', 'West']
]

index = pd.MultiIndex.from_arrays(arrays, names=('Country', 'Region'))

df = pd.DataFrame({'Sales': [100, 150, 200, 250]}, index=index)

print(df.loc[('Canada', 'East')])
```

```python
import pandas as pd

df = pd.DataFrame({
    'Department': ['HR', 'HR', 'IT', 'IT'],
    'Gender': ['Male', 'Female', 'Male', 'Female'],
    'Salary': [50000, 52000, 70000, 68000]
})

pivot_df = df.pivot_table(values='Salary', index='Department', columns='Gender', aggfunc='mean')

print(pivot_df)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Values': [10, 20, 30, 40, 50]
})

df['Rolling_Mean'] = df['Values'].rolling(window=2).mean()

print(df)
```
