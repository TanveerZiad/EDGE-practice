```python
items = ["Apples", "Bananas", "Oranges", "Bread", "Milk"]

with open("shopping_list.txt", "w") as file:
    for item in items:
        file.write(item + ":\n")
```

```python
with open("shopping_list.txt", "r") as file:
    for line_number, line in enumerate(file, start=1):
        item = line.strip()
        print(f"{line_number}. {item}")
```

```python
with open("shopping_list.txt", "a") as file:
    file.write("Eggs:\n")
with open("shopping_list.txt", "r") as file:
    lines = file.readlines()
for index, line in enumerate(lines, start=1):
    item = line.strip()
    print(f"{index}. {item}")
```

```python
try:
    with open("grocery_list.txt", "r") as file:
        lines = file.readlines()
        for line in lines:
            print(line.strip())
except FileNotFoundError:

    print("Error: The file 'grocery_list.txt' was not found.")
```

```python
word_count = 0
try:

    with open("shopping_list.txt", "r") as file:

        for line in file:

            words = line.split()
            word_count += len(words)

    print(f"Total number of words (items) in shopping_list.txt: {word_count}")

except FileNotFoundError:
    print("Error: The file 'shopping_list.txt' was not found.")
```

```python
lines = [
    "Python is great for data analysis.",
    "Exception handling makes code robust.",
    "File handling is crucial for data storage."
]
with open("notes.txt", "w") as file:
    for line in lines:
        file.write(line + "\n")
print("Notes written to notes.txt successfully.")
```

```python
with open("notes.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
try:
    with open("notes.txt", "r") as source_file:

        content = source_file.read()


    with open("notes_backup.txt", "w") as backup_file:

        backup_file.write(content)

    print("File contents copied to notes_backup.txt successfully.")

except FileNotFoundError:
    print("Error: The file 'notes.txt' was not found.")

except IOError:
    print("Error: An I/O error occurred during file operations.")
```
