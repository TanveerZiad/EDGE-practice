```python
def main():
    try:
        expression = input("Enter an arithmetic expression: ")
        result = eval(expression)
        print("The result of the expression is:", result)
    except (SyntaxError, NameError, ZeroDivisionError) as e:
        print("Error: Invalid expression -", e)

if __name__ == "__main__":
    main()
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print(f"{num1} < {num2}: {num1 < num2}")
print(f"{num1} > {num2}: {num1 > num2}")
print(f"{num1} == {num2}: {num1 == num2}")
print(f"{num1} != {num2}: {num1 != num2}")
print(f"{num1} <= {num2}: {num1 <= num2}")
print(f"{num1} >= {num2}: {num1 >= num2}")
```

```python
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(f"{num} is not a prime number.")
            break
    else:
        print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
import random

secret_number = random.randint(1, 100)
attempts = 10
print("Guess the secret number between 1 and 100!")

while attempts > 0:
    guess = int(input(f"You have {attempts} attempts left. Enter your guess: "))
    if guess < secret_number:
        print("Too low!")
    elif guess > secret_number:
        print("Too high!")
    else:
        print("Congratulations! You guessed the right number.")
        break
    attempts -= 1

if attempts == 0:
    print(f"Sorry, you have no attempts left. The secret number was {secret_number}.")
```

```python
from statistics import mean, median, mode

numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))

print("Mean:", mean(numbers))
print("Median:", median(numbers))
print("Mode:", mode(numbers))
```

```python
list1 = list(map(int, input("Enter first list of numbers separated by spaces: ").split()))
list2 = list(map(int, input("Enter second list of numbers separated by spaces: ").split()))

union = list(set(list1) | set(list2))
intersection = list(set(list1) & set(list2))
symmetric_difference = list(set(list1) ^ set(list2))

print("Union:", union)
print("Intersection:", intersection)
print("Symmetric Difference:", symmetric_difference)
```

```python
tuples = eval(input("Enter a list of tuples: "))

dict_from_tuples = dict(tuples)
sorted_dict = dict(sorted(dict_from_tuples.items()))

print("Sorted Dictionary:", sorted_dict)
```

```python
from collections import Counter

# Read the file
with open('input.txt', 'r') as file:
    text = file.read()

# Count word frequencies
words = text.split()
word_count = Counter(words)

# Save the results to another file
with open('output.txt', 'w') as output_file:
    for word, count in word_count.items():
        output_file.write(f"{word}: {count}\n")

print("Word frequency has been saved to 'output.txt'.")
```
